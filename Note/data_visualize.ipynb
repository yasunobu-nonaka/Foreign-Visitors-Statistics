## 年ごとの訪日外客数総数をまとめる


```python
# 年ごとに適切な列インデックスを使用してデータを取得
def extract_total_visitors_by_year():
    """
    各シートの総数列（1月から累計、伸率まで）を抽出し、
    年ごとの訪日外客数の総数をまとめたDataFrameを作成
    2019年までと2020年移行で列構造が異なるため、年ごとに適切な列インデックスを使用
    """
    # 全シートを読み込み
    all_sheets = pd.read_excel(data_path, sheet_name=None, header=3)
    
    # 年ごとのデータを格納するリスト
    yearly_data = []
    
    # 各シートを処理
    for year, df in all_sheets.items():
        try:
            # 総数行を取得（より堅牢な方法）
            total_row = df[df.iloc[:, 0].astype(str).str.contains('総数', na=False)]
            
            if not total_row.empty:
                # 年を整数に変換
                year_int = int(year)
                
                # 2019年までと2020年移行で列インデックスを調整
                if year_int <= 2019:
                    # 2019年まで: 1月は列インデックス1から開始
                    start_col = 1
                else:
                    # 2020年移行: 1月は列インデックス2から開始
                    start_col = 2
                
                # 1月から累計、伸率までの列を取得
                # 列構造: 1月, 伸率, 2月, 伸率.1, ..., 12月, 伸率.11, 累計, 伸率.12
                end_col = start_col + 26  # 26列分（1月から伸率.12まで）
                monthly_data = total_row.iloc[0, start_col:end_col].values
                
                # データを辞書として保存
                year_dict = {'年': year_int}
                
                # 月別データと伸率を追加
                months = ['1月', '2月', '3月', '4月', '5月', '6月', 
                         '7月', '8月', '9月', '10月', '11月', '12月']
                growth_rates = ['伸率', '伸率.1', '伸率.2', '伸率.3', '伸率.4', '伸率.5',
                              '伸率.6', '伸率.7', '伸率.8', '伸率.9', '伸率.10', '伸率.11']
                
                # 月別データを追加（月と伸率が交互に並んでいる）
                for i, month in enumerate(months):
                    month_idx = i * 2  # 月のインデックス（0, 2, 4, ...）
                    growth_idx = i * 2 + 1  # 伸率のインデックス（1, 3, 5, ...）
                    
                    if month_idx < len(monthly_data):
                        year_dict[month] = monthly_data[month_idx] if pd.notna(monthly_data[month_idx]) else None
                    if growth_idx < len(monthly_data):
                        year_dict[growth_rates[i]] = monthly_data[growth_idx] if pd.notna(monthly_data[growth_idx]) else None
                
                # 累計と伸率.12を追加（最後の2列）
                if len(monthly_data) >= 25:  # 累計のインデックス
                    year_dict['累計'] = monthly_data[24] if pd.notna(monthly_data[24]) else None
                if len(monthly_data) >= 26:  # 伸率.12のインデックス
                    year_dict['伸率.12'] = monthly_data[25] if pd.notna(monthly_data[25]) else None
                
                yearly_data.append(year_dict)
                
        except Exception as e:
            print(f"年 {year} の処理でエラーが発生しました: {e}")
            continue
    
    # DataFrameを作成
    yearly_df = pd.DataFrame(yearly_data)
    
    # 年でソート
    yearly_df = yearly_df.sort_values('年').reset_index(drop=True)
    
    return yearly_df

# データフレームを作成
yearly_visitors_df = extract_total_visitors_by_year()
print("年ごとの訪日外客数総数データフレーム:")
print(yearly_visitors_df.head(10))
```

```python
# 年ごとの年間総数を計算
def calculate_yearly_totals(df):
    """
    年ごとの年間総数を計算
    累計列がある場合はそれを使用、ない場合は月別データを合計
    """
    yearly_totals = []
    
    for _, row in df.iterrows():
        year = row['年']
        
        # まず累計列を確認
        if '累計' in row and pd.notna(row['累計']):
            yearly_total = row['累計']
        else:
            # 累計がない場合は月別データを合計
            monthly_values = []
            months = ['1月', '2月', '3月', '4月', '5月', '6月', 
                     '7月', '8月', '9月', '10月', '11月', '12月']
            
            for month in months:
                if month in row and pd.notna(row[month]):
                    monthly_values.append(row[month])
            
            # 年間総数を計算
            if monthly_values:
                yearly_total = sum(monthly_values)
            else:
                yearly_total = None
        
        yearly_totals.append({'年': year, '年間総数': yearly_total})
    
    return pd.DataFrame(yearly_totals)

# 年間総数を計算
yearly_totals_df = calculate_yearly_totals(yearly_visitors_df)
print("年ごとの年間総数:")
print(yearly_totals_df)
```

```python
# 年ごとの訪日外客数の推移を可視化
plt.figure(figsize=(14, 8))

# データが存在する年のみをプロット
valid_data = yearly_totals_df.dropna()
plt.plot(valid_data['年'], valid_data['年間総数'], marker='o', linewidth=2, markersize=6)

plt.title('年ごとの訪日外客数推移', fontsize=16)
plt.xlabel('年', fontsize=12)
plt.ylabel('訪日外客数', fontsize=12)
plt.grid(True, alpha=0.3)

# y軸のフォーマットを設定
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().ticklabel_format(style='sci', axis='y', scilimits=(0, 0))

# x軸の目盛りを設定
plt.xticks(rotation=45)

# 2020年を境に色分け（オプション）
plt.axvline(x=2020, color='red', linestyle='--', alpha=0.7, label='2020年（コロナ影響）')
plt.legend()

plt.tight_layout()
plt.show()
```
