# Visualize Data

- 使用データ: 日本政府観光局(JNTO), 訪日外客統計, https://www.jnto.go.jp/statistics/data/visitors-statistics/

## packages, variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

plt.rcParams['font.family'] = 'Noto Sans CJK JP'
```

```python
data_path = "../Data/monthly-foreign-visitors/20250820_1615-5.xlsx"

months_of_year = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
```

## load data

```python
df = pd.read_excel(data_path, sheet_name="2025", header=3)
```

```python
df.head()
```

```python
df.columns
```

```python
# replce "Unnamed" column names
df.columns = ['country_col1', 'country_col2', '1月', '伸率', '2月', '伸率.1', '3月', '伸率.2',
       '4月', '伸率.3', '5月', '伸率.4', '6月', '伸率.5', '7月', '伸率.6', '8月', '伸率.7',
       '9月', '伸率.8', '10月', '伸率.9', '11月', '伸率.10', '12月', '伸率.11', '累計',
       '伸率.12']
```

```python
df.info()
```

```python
row = df[df['country_col1'] == '総数']
row_series = row.iloc[0, 2:][months_of_year]

ax = row_series.plot(kind="bar")

ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))

plt.ylabel('訪日外客数')
plt.title('訪日外客数（2025年）')

plt.show()
```
